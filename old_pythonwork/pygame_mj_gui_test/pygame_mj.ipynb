```python
import pygame
from pygame.locals import *
from sys import exit
background_image_filename = 'green_background.png'
mj_tile = []
mj_tile_c =[]
for i in range (0,34):
    mj_tile.append(str(i)+'.png')
    
    mj_tile_c.append(pygame.image.load(mj_tile[i]))
    mj_tile_c[i]= pygame.transform.scale(mj_tile_c[i], (50, 70))


pygame.init()
screen = pygame.display.set_mode((1000, 720), 0, 32)
background = pygame.image.load(background_image_filename).convert()
#mj_tile_c = pygame.image.load(mj_tile[2]).convert_alpha()


x, y = 0, 0
move_x, move_y = 0, 0



while True:
    
    for event in pygame.event.get():
        if event.type == QUIT:
            exit()
        if event.type == KEYDOWN:
            #键盘有按下？
            if event.key == K_LEFT:
                #按下的是左方向键的话，把x坐标减一
                move_x = -10
            elif event.key == K_RIGHT:
                #右方向键则加一
                move_x = 10
            elif event.key == K_UP:
                #类似了
                move_y = -10
            elif event.key == K_DOWN:
                move_y = 10
            elif event.key == K_q:
                pygame.display.quit()
        elif event.type == KEYUP:
            #如果用户放开了键盘，图就不要动了
            move_x = 0
            move_y = 0

        #计算出新的坐标
        x+= move_x
        y+= move_y

      
        screen.blit(background, (0,0))

        screen.blit(mj_tile_c[5], (x,y))
        x_tile=30
        for i in range (0,17):    
            if(playerlist[0][i]==42):
                continue
            screen.blit(mj_tile_c[playerlist[0][i]], (x_tile,600))
            x_tile+=55
        



        #在新的位置上画图
        pygame.display.update()
```
